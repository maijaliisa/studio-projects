# scraping NPR Morning Edition 'Atlanta' search 


```python
response = requests.get('http://www.npr.org/templates/search/index.php?searchinput=Atlanta&programId=3&tabId=all&dateId=0&sort=match&sort=match&start=0')
me_doc = BeautifulSoup(response.text, 'html.parser')
```

```python
headlines = me_doc.find_all('h1')
teasers = me_doc.find_all('div', { 'class': 'storyinfo' }) 

for headline in headlines:
    print(headline.text) 
    print('---------')
    if headline:
        print(teasers) 
```
