```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
#STEP ONE - FIND PRIORITY RECOMMENDATIONS BY AREA - PRIORITY_GAO_RECS: https://www.gao.gov/resources/presidential-transition/priority-recommendations?browse=topic_ss#results
```

```python
areas = pd.read_csv('PRIORITY_GAO_RECS - open_priorities_by_topic.csv')
areas_sorted = areas.sort_values(by='Count',ascending=True)
areas_sorted.plot(kind='barh',x='Open Priorities',legend=False, title='Open Priority Items By Topic (As of June 2017)')
plt.savefig('All_Open_Priorities.pdf')
```

```python
areas_sorted
```

```python
areas.sort_values(by='Count',ascending=False).head(10).plot(kind='bar',x='Open Priorities', figsize=(15,4),legend=False,title='Open Priority Recommendations By Topic')
plt.savefig('Open Priorities By Topic.pdf')
```
