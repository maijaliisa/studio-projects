```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('GAORecommendations - TRUE.csv')
df.head()
```

```python
df_cyber = df[df['Topic'] == 'Cybersecurity']
df_cyber.shape
```

```python
#matplotlib.rcParams['pdf.fonttype'] = 42
```

```python
df_cyber.sort_values(by='Count', ascending=True).plot(x='Department',y='Count',kind='barh',figsize=(14,8),legend=False,title='Open Cybersecruity Priority Recommendations')
plt.savefig('OpenCyberSecurity.svg')
```

```python
## 68 priority items are cybersecurity 
```

```python
df_gov = df[df['Topic'] == 'Government Operations']
df_gov.shape
```

```python
df_gov.sort_values(by='Count', ascending=True).plot(x='Department',y='Count',kind='barh',figsize=(14,8),legend=False, title="Open Government Operations Priority Items")
plt.savefig('OpenGovernmentOperations.svg')
```
