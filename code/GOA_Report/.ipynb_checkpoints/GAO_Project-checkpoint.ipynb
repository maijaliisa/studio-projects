```python
import requests
from bs4 import BeautifulSoup
```

```python
#The URL for Legislative GAO reports 01/02/2009 - 07/24/2017
url = 'http://www.gao.gov/browse/agency/Legislative/custom?&rows=50&o=100&now_sort=issue_date_dt+desc%2Ctitle_sort+asc&adv_begin_date=01/03/2009&adv_end_date=07/24/2017&o=150'
response = requests.post(url)
doc = BeautifulSoup(response.text, 'html.parser')  
#doc.prettify()  
```

scraping the first page 

```python
#first find div 
first_div = doc.find('div', {'class':'inner'})
first_div
```

```python
# https://stackoverflow.com/questions/5041008/how-to-find-elements-by-class
# https://cha.house.gov/sites/republicans.cha.house.gov/files/documents/ALDTC_Data_Analytics_GAO_NabaBarkakati_panel5.pdf
# https://www.crummy.com/software/BeautifulSoup/bs4/doc/
# https://stackoverflow.com/questions/40218812/how-to-scrap-span-ids-texts-in-beautifulsoup-in-the-following-html
# http://jonathansoma.com/lede/foundations-2017/classes/adv-scraping/scraping-bbc/ 
```

```python
soup_doc = first_div.find_all('div', {'class': 'listing grayBorderTop'})
soup_doc
```

Exporting the first page to CSV to clean up

```python
big_list = []
reports = first_div.find_all('span')
for rep in reports:
    big_dict = {}
    dates = rep.find('span')
    if dates:
        big_dict['dates'] = dates.text
    big_list.append(big_dict)
import pandas as pd
df = pd.DataFrame(big_list) 
df.to_csv("PAGE444444.csv", index=False) 
```

```python
next_list = []
titles = first_div.find_all('a')
for title in titles:
    next_dict = {}
    titles = title.find('span')
    if titles:
        next_dict['titles'] = titles.text
    next_list.append(next_dict)
df2 = pd.DataFrame(next_list)
df2.to_csv("PAGE44444TITLES.csv", index= False)         
```
