```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

Goal: As film/TV industry grows in Atlanta, look at how arts/culture is also growing in the State 

```python
#Arts and Culture Compensation and Employment by State - found https://www.bea.gov/regional/downloadzip.cfm 
ga_arts = pd.read_csv('GA_In_The_Arts_Employment_True2.csv') 
#ga_arts['Description'].astype(str)
#ga_arts['Description'].apply(lambda x:)
ga_arts.set_index('Description', inplace=True)
ga_arts
```

```python
#ga_arts['Description'].strip()
```

```python
ax = ga_arts.loc['Agents/Managers For Artists'].plot(figsize=(15,5),legend=True)
#ga_arts.loc['Motion Pictures'].plot(figsize=(15,5), legend=True, ax=ax)
ga_arts.loc['Independent Artists, Writers, And Performers'].plot(figsize=(15,5), legend=True, ax=ax, title='Breaking Down Film Talent In Atlanta')
ga_arts.loc['Fine Arts Education'].plot(figsize=(15,5), legend=True, ax=ax)
ga_arts.loc['Motion Pictures'].plot(figsize=(15,5), legend=True, ax=ax)

```

```python
#Make a chart of: "Agents/Managers For Artists, Motion Pictures, Sound Recording 
#This chart will show how film is driving specific jobs" 
cols_of_interest = ['2009','2010','2011','2012','2013','2014']
new_df = ga_arts[cols_of_interest]
new_df
```
