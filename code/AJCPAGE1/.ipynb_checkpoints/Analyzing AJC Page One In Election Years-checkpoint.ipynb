```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
df2011 = pd.read_csv('Local News Week 3 - Page1Subjects2011.csv')
df2011.rename(columns={'2011':'Subject'}, inplace=True)
df2011.head()
```

```python
df2012 = pd.read_csv('Local News Week 3 - Page1Subjects2012.csv')
df2012.rename(columns={'Document Count': '2012count'}, inplace=True)
```

```python
merged2012 = pd.merge(df2011,df2012,on='Subject',how='outer')
merged2012.sort_values(by='2012count',ascending=False)
```

```python
############################################################################################################################################
######2016 vs. 2015######
```

```python
df2015 = pd.read_csv('Local News Week 3 - Page1Subjects2015.csv')
df2015.rename(columns={'2015':'Subject'}, inplace=True)
df2015
```

```python
df2016 = pd.read_csv('Local News Week 3 - Page1Subjects2016.csv')
df2016.rename(columns={'2016':'Subject'}, inplace=True)
df2016
```

```python
dfnew = pd.merge(df2015,df2016,on='Subject',how='outer')
dfnew.dropna(inplace=True)
```

```python
dfnew.sort_values(by='2016count',ascending=False)
```
