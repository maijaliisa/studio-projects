```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
df2011 = pd.read_csv('Local News Week 3 - Page1Subjects2011 (1).csv')
#df2011.set_index('Subject', inplace=True)
```

```python
df2012 = pd.read_csv('Local News Week 3 - Page1Subjects2012 (2).csv')
#df2012.set_index('Subject', inplace=True)
```

```python
df2013 = pd.read_csv('Local News Week 3 - Page1Subjects2013 (1).csv')
#df2013.set_index('Subject', inplace=True)
```

```python
df2014 = pd.read_csv('Local News Week 3 - Page1Subject2014 (1).csv')
#df2014.set_index('Subject', inplace=True)
```

```python
df2015 = pd.read_csv('Local News Week 3 - Page1Subjects2015 (1).csv')
#df2015.set_index('Subject',inplace=True)
```

```python
df2016 = pd.read_csv('Local News Week 3 - Page1Subjects2016 (1).csv')
df2016.rename(columns={'Subjecty':'Subject'}, inplace=True)
```

```python
#df2016.set_index('Subject',inplace=True)
```

```python
all_merged = df2011.merge(df2012,on='Subject',how='outer').merge(df2013,on='Subject',how='outer').merge(df2014,on='Subject',how='outer').merge(df2015, on='Subject', how='outer').merge(df2016,on='Subject', how='outer')
all_merged.sort_values(by='2016count',ascending=False)
```

```python
all_merged.set_index('Subject',inplace=True)
```

```python
all_merged.sort_values(by='2016count',ascending=False)
```

```python
del all_merged['2011count']
```

```python
ax = all_merged.loc['Domestic Politics'].plot(figsize=(16,6),legend=True, title='AJC Front Page Politics Topics')
all_merged.loc['Regional Politics'].plot(legend=True)
plt.savefig('AJC FRONT PAGE Politics Topics.svg')
```
