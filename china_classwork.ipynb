```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt

import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
```

```python
df = pd.read_csv('nobel_prize - Sheet1 (2).csv')
df
```

```python
df2 = df.drop(df.index[[23,24,25,26,27,28,29]])
df2
```

```python
ax = df2.sort_values(by='Nobel Prize').plot(figsize=(7,10), x='Country',y="Nobel Prize", kind='barh', color='#d3d4d5', title='NUMBER OF NOBEL PRIZES')
ax.patches[df2.index[2]].set_facecolor('r')
plt.savefig("nobel_prize_graph.pdf") 
```
